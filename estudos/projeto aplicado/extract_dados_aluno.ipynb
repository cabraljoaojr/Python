```python
# importando biblioteca
import pg8000
import pandas as pd

# Configurar conexão com banco local
conn = pg8000.connect(
    host="localhost",
    database="educacao",
    user="postgres",
    password="admin",
    port=5432
)

# Query para extrair dados da view
query = "SELECT * FROM educacao.view_aluno_turma_detalhada"

# Executar a query e criar DataFrame
try:
    cursor = conn.cursor()
    cursor.execute(query)
    dados = cursor.fetchall()
    col_names = [desc[0] for desc in cursor.description]  # Obter nomes das colunas
    
    df = pd.DataFrame(dados, columns=col_names)
    
finally:
    conn.close()
    
# Visualizando DataFrame
df
```
